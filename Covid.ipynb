## Пример 1

##### D4 - dataframe

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
model = KMeans(n_clusters=6, random_state = 13)
model.fit(D4)
y_pred = model.predict(D4)
```

```python
DopDF = pd.DataFrame(columns=['cluster'])
DopDF['cluster'] = y_pred
```

```python
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(D4)
principalDf = pd.DataFrame(data = principalComponents, columns = ['principal component 1', 'principal component 2'])
principalDf.head(5)
```

```python
finalDf = pd.concat([principalDf, DopDF['cluster']], axis = 1)
finalDf.head(5)
```

```python
fig = plt.figure(figsize = (10,10))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 Component PCA', fontsize = 20)


targets = [0, 1, 2, 3, 4, 5]
colors = ['#00ffff', '#ff00ff', '#0000ff','#00ff00','#ffff00','#ff0000']
for target, color in zip(targets,colors):
    indicesToKeep = finalDf['cluster'] == target
    ax.scatter(finalDf.loc[indicesToKeep, 'principal component 1']
               , finalDf.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.decomposition import PCA
```

```python
clusters_kmeans = KMeans(n_clusters=6, random_state=777).fit_predict(X)
clusters_hierar = AgglomerativeClustering(n_clusters=6, linkage='ward').fit_predict(X)
clusters_dbscan = DBSCAN(eps=0.3, min_samples=2).fit_predict(X)
```

```python
pca = PCA(n_components=2) 
points = pca.fit_transform(X) 
# доля отклонения, объясняемый каждым из выбранных компонентов
print(pca.explained_variance_ratio_) 
```

```python
x, y = zip(*points)
plt.scatter(x, y, c=clusters_kmeans)
plt.scatter(x, y, c=clusters_hierar)
plt.scatter(x, y, c=clusters_dbscan)
```
