### Описание столбцов 

Facility.Name - Название госпиталя

Facility.City - город, в котором находится госпиталь

Facility.State - Аббревиатура штата, в котором расположена больница

Facility.Type - Тип организации, управляющей больницей: Государственная, Частная, На имущественных правах, Церковная или 
Неизвестная

Rating.Overall - Общая оценка от 1 до 5 звезд, причем 5 звезд являются наивысшей оценкой; -1 означает отсутствие рейтинга

Rating.Mortality - Значения выше, также, ниже или неизвестно в сравнении со смертностью с нац.больницой

Rating.Safety - Значения выше, также, ниже или неизвестно в сравнении с безопасностью с нац.больницой

Rating.Readmission - Значения выше, также, ниже или неизвестно в сравнении с реадмиссией с национальной больницой

Rating.Experience - Значения выше, также, ниже или неизвестно в сравнении с уровнем терпимости с национальной больницой

Rating.Effectiveness - Значения выше, также, ниже или неизвестно в сравнении с эффективностью медицинской помощи

Rating.Timeliness - Значения выше, также, ниже или неизвестно в сравнении с своевременностью оказания медицинской помощи с национальной больницей 

Rating.Imaging - Значения выше, также, ниже или неизвестно в сравнении с эффективным использование визуализации с национальной
больницой

Procedure.Heart Attack.Cost - Средняя стоимость лечения сердечных приступов с национальной больницой

Procedure.Heart Attack.Quality - Значения выше, также, ниже или неизвестно в сравнении с качеством медицинской помощи при сердечных приступах с национальной больницой

Procedure.Heart Attack.Value - Значения выше, также, ниже или неизвестно в сравнении со стоимостью лечения сердечных приступов с национальной больницой

Procedure.Heart Failure.Cost - Средняя стоимость лечения сердечной недостаточности

Procedure.Heart Failure.Quality - Значения выше, также, ниже или неизвестно в сравнении с качеством медицинской помощи при сердечной недостаточности с национальной больницой

Procedure.Heart Failure.Value - Значения выше, также, ниже или неизвестно в сравнении со стоимостью лечения сердечной недостаточности с национальной больницой

Procedure.Pneumonia.Cost - Средняя стоимость лечения пневмонии

Procedure.Pneumonia.Quality - Значения выше, также, ниже или неизвестно в сравнении с качеством медицинской помощи при пневмонии с национальной больницой

Procedure.Pneumonia.Value - Значения выше, также, ниже или неизвестно в сравнении со стоимостью лечения пневмонии с национальной больницой

Procedure.Hip Knee.Cost - Средняя стоимость лечения заболеваний тазобедренного или коленного сустава

Procedure.Hip Knee.Quality - Значения выше, также, ниже или неизвестно в сравнении с качеством медицинской помощи при 
заболеваниях тазобедренного сустава или колена с национальной больницой

Procedure.Hip Knee.Value - Значения выше, также, ниже или неизвестно в сравнении со стоимостью лечения заболеваний 
тазобедренного или коленного сустава с национальной больницей
