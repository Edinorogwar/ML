### Предварительная обработка данных   

#### Стандартизация или удаление среднего и масштабирование дисперсии

```python
import pandas as pd 
import matplotlib as plt
import warnings
warnings.simplefilter('ignore')
```

```python
data = pd.DataFrame(
    {
        'Lucky':[50, 70, 38, 29],
        'Rating':[24, 35, 12, 0],
        'Friendly':[0.8, 0.2, 0.99, 0.01],
        'Duration':[2, 3, 0.5, 1],
        'Sneakers':['Yes', 'Yes', 'No', 'Yes'],
        'Vest':['Yes', 'No', 'Yes', 'Yes'],
    }
)
data
```

### Применяем LabelEncdore, чтобы перевести категориальные признаки в числовые

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df = data.copy(deep=True)
```

```python
Sneakers_le = LabelEncoder()
df.Sneakers = Sneakers_le.fit_transform(df.Sneakers)

Vest_le = LabelEncoder()
df.Vest = Vest_le.fit_transform(df.Vest)

df
```
