```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
plt.style.use('seaborn-whitegrid')
```

```python
#pd.set_option("display.max.columns", None)
df = pd.read_csv('https://raw.githubusercontent.com/owid/covid-19-data/master/public/data/owid-covid-data.csv', delimiter =',')
df.head()
```

```python
d = df.count()
d
```

```python
d = df.groupby('continent')['life_expectancy'].max()
d
```

```python
d1 = df.groupby('continent')['total_cases'].sum()
d1
```

```python
fig, ax = plt.subplots()
ax.plot(d)
ax.legend(['life_expectancy'])
```

```python
fig, ax = plt.subplots()
ax.plot(d1)
ax.legend(['total_cases'])
```
